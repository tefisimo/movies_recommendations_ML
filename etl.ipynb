#### Continuacion de la fase 2 ahora con el dataset "Credits.csv"

```python
credits.head()
```

```python
df_cast = pd.DataFrame()
df_cast['cast'] = credits['cast']
```

```python
df_cast
```

```python
# Confirmamos y transformamos a diccionarios a partir de 'str'
df_cast['cast'] = df_cast['cast'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

# create new columns from dict
df_cast['cast_id'] = df_cast['cast'].apply(lambda x: [d['cast_id'] for d in x] if isinstance(x, list) else [])
df_cast['character'] = df_cast['cast'].apply(lambda x: [d['character'] for d in x] if isinstance(x, list) else [])
df_cast['credit_id'] = df_cast['cast'].apply(lambda x: [d['credit_id'] for d in x] if isinstance(x, list) else [])
df_cast['actor_id'] = df_cast['cast'].apply(lambda x: [d['id'] for d in x] if isinstance(x, list) else [])
df_cast['actor_name'] = df_cast['cast'].apply(lambda x: [d['name'] for d in x] if isinstance(x, list) else [])
```

```python
df_cast.drop(columns=['cast', 'cast_id', 'credit_id'], inplace=True)
```

```python
df_crew = pd.DataFrame()
df_crew['crew'] = credits['crew']
```

```python
df_crew.head(1)
```

```python
# check and transform str to dict
df_crew['crew'] = df_crew['crew'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

# Verificamos en la lista de diccionarios en busqueda del id y procedemos a crear una columna con los pertenecientes por fila
df_crew['crew_id'] = df_crew['crew'].apply(lambda x: [d['id'] for d in x] if isinstance(x, list) else [])
df_crew['job'] = df_crew['crew'].apply(lambda x: [d['job'] for d in x] if isinstance(x, list) else [])
df_crew['crew_name'] = df_crew['crew'].apply(lambda x: [d['name'] for d in x] if isinstance(x, list) else [])
```

```python
df_crew.head(1)
```

```python
df_crew['job'][0]
```

```python
df_crew['crew_name'][0]
```

```python
# Define una funcion para extraer el director de la pelicula
def extraer_director(row):
    """Esta funcion extrae el director en un rango de una lista"""
    lista_trabajos = row['job']
    lista_nombres = row['crew_name']
    directors = []  
    for i in range(len(lista_trabajos)):
        if lista_trabajos[i] == 'Director':
            directors.append(lista_nombres[i])
    return ', '.join(directors)  # retorna nombre de los directores separados por comas si existe mas de 1
```

```python
# Aplicamos la funcion al dataframe
df_crew['Director'] = df_crew.apply(extraer_director, axis=1)
```

```python
df_crew['Director']
```
